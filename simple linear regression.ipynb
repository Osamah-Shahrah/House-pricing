```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv("kc_house_data.csv")
print(data.info())
```

```python
space=data['sqft_living']
price=data['price']
x=np.array(space).reshape(-1,1)
y=np.array(price)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=1/3,random_state=0)
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
regression.fit(xtrain,ytrain)
pred=regression.predict(xtest)
```

```python
plt.scatter(xtrain,ytrain,color="red")
plt.plot(xtrain,regression.predict(xtrain), color= "blue")
plt.title("Visuals the train ")
plt.xlabel("space")
plt.ylabel("price")
plt.show()


plt.scatter(xtest,ytest,color="red")
plt.plot(xtest,regression.predict(xtest),color="blue")
plt.title("Visuals the Test")
plt.xlabel("space")
plt.ylabel("price")
plt.show()
```
